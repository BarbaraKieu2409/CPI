```python
import pandas as pd 
df_CPI = pd.read_csv('D:/project_data_mining/dataset_CPI.csv')
# obs_value: observation value 
# dữ liệu lấy từ 01-2011-> 12-2025


# # Chuyển TIME_PERIOD sang định dạng YYYY-MM (nếu đang ở dạng "1971M01")
df_CPI['TIME_PERIOD'] = df_CPI['TIME_PERIOD'].str.replace('M', '', regex=False)  
df_CPI = df_CPI[['COUNTRY','COICOP_1999', 'TIME_PERIOD', 'OBS_VALUE']].copy()
df_CPI = df_CPI.rename(columns={'COICOP_1999': 'Expenditure_category', 'OBS_VALUE': 'CPI'})
df_CPI.head(5)
```

```python
import pandas as pd

df2 = pd.read_excel('D:/project_data_mining/CMO-Historical-Data-Monthly.xlsx',sheet_name="Monthly Indices", header=5 , engine='openpyxl')
# df2.head(5)

df2 = df2.rename(columns={df2.columns[0]: "TIME_PERIOD"})

# change name column 
df_energy = df2[["TIME_PERIOD", "Energy"]].copy()

df_energy["TIME_PERIOD"] = (
    df_energy["TIME_PERIOD"]
    .astype(str)
    .str.replace("M", "-", regex=False)
)
# convert TIME_PERIOD to YYYY-MM format 
df_energy["TIME_PERIOD"] = pd.to_datetime(df_energy["TIME_PERIOD"],format="%Y-%m")
df_energy["TIME_PERIOD"] = df_energy["TIME_PERIOD"].dt.strftime("%Y-%m")

# filter 2000–2024
df_energy = df_energy[(df_energy["TIME_PERIOD"] >= "2000-01") & (df_energy["TIME_PERIOD"] <= "2025-03")]

# save CSV file
# df_energy.to_csv("energy_monthly_2000_2024.csv", index=False)

df_energy.tail()
```

```python
# merge energy and CPI dataset  
df = pd.merge(
    df_CPI,
    df_energy,
    on="TIME_PERIOD",
    how="left"   )
df.tail(100)
```

```python
import pandas as pd

df_GDP = pd.read_csv('D:/project_data_mining/GDP_dataset.csv')

# change name column time_period to year
df_GDP = df_GDP.rename(columns={'TIME_PERIOD': 'YEAR', 'OBS_VALUE': 'GDP'})
# df_GDP.columns
df_GDP['YEAR'] = df_GDP['YEAR'].astype(int)

# get year from time_period month 
df['YEAR'] = df['TIME_PERIOD'].str.slice(0, 4)
df['YEAR'] = pd.to_numeric(df['YEAR'], errors='coerce') 
df['YEAR'] = df['YEAR'].astype('Int64')

# merge data with GDP dataset 
df = df.merge(df_GDP[['COUNTRY', 'YEAR', 'GDP']], on =['COUNTRY', 'YEAR'], how='left')
df.head(5)
```
